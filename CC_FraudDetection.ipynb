DISADVANTAGE OF UNDERSAMPLING IS THAT WE LOSE A LOT OF VALUABLE DATA
THAT'S WHY WE ARE GONNA USE OVERSAMPLING

**OVERSAMPLING**

WE DO OVERSAMPLING USING **Synthetic Minority Oversampling Technique (SMOTE) TECHNIQUE**
THIS TECHNIQUE MAINTAINS CLASS DISTRIBUTION **BY RANDOMLY INCREASING MINORITY CLASS EXAMPLES BY** **REPLICATING THEM**
HERE YOU ARE NOT CREATING DUPLICATES RATHER YOU ARE CREATING SYNTHETIC DATA POINTS

```python
x= data.drop('Class', axis=1)
y=data['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res= SMOTE().fit_resample(x,y)
```

```python
y_res.shape
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test= train_test_split(x_res,y_res,test_size=0.2, random_state=42)
```
